```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 3*x**2
```

```python
x = []
y = []
for i in range(20):
    x.append(i)
    y.append(f(i))
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,200,10)
y = f(x)
y
```

```python
plt.plot(x,y)
plt.title("Graf $3x^2$")
plt.style.use("fivethirtyeight")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
