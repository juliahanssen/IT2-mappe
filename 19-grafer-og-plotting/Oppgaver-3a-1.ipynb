```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Oppagve 3 
def f(x):
    return 2 * x - 3

x = np.linspace(0, 50, 100)
y = f(x)

plt.plot(x,y, color="coral")
plt.title("Oppgave 3")
plt.style.use("fivethirtyeight")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid()

plt.show()
```

```python
## Oppgave 4

xverdier = np.linspace(0, 20, 50)

# Graf 1
yverdier = 0.5*xverdier**2 

plt.subplot(1, 2, 1)
plt.plot(xverdier, yverdier)
plt.grid()

# Graf 2
yverdier = -0.3*xverdier**3 

plt.subplot(1, 2, 2)
plt.plot(xverdier, yverdier)
plt.grid()

plt.show()
```
