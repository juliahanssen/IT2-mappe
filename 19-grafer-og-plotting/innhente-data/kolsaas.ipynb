```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linjer = linje.split(",")
    tur.append(splittet_linjer)
tur[0][1]
```

```python
x = []
y = []
lengde = []
bredde = []

for linje in tur[1:]:
    tid = float(linje[0])
    hoyde = float(linje[3])
    lengde.append(float(linje[2]))
    bredde.append(float(linje[1]))
    x.append(tid)
    y.append(hoyde)


#for målepunkt in range(1, len(tur)):
    # x.append(tur[målepunkt][0])
    # y.append(tur[målepunkt][3])

plt.scatter(x,y)
plt.style.use("fivethirtyeight")
plt.show()
```
